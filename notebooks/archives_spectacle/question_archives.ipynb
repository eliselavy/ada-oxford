```python
# 1744 Pôle sud
# 439 Arsenal
# voir si regroupement avec les spectaclesId

import pandas as pd

csv_file_path_pole_sud = "all_10_years_organism_1744_convertInt.csv"
csv_file_path_arsenal = "all_10_years_organism_439_convertInt.csv"


df_pole_sud = pd.read_csv(csv_file_path_pole_sud)
# drop=True Delete columns to be used as the new index.
# inplace=True Modify the DataFrame in place (do not create a new object).
# verify_integrity Check the new index for duplicates. Otherwise defer the check until necessary. Setting to False will improve the performance of this method.

df_pole_sud= df_pole_sud[df_pole_sud['performanceId'] != 0]
#df_pole_sud.set_index('performanceId', drop=True, inplace=True)

#['year', 'month']

df_pole_sud.set_index(['performanceId', 'year'], drop=True, inplace=True)

df_arsenal = pd.read_csv(csv_file_path_arsenal)
df_arsenal= df_arsenal[df_arsenal['performanceId'] != 0]

#df_arsenal.set_index('performanceId', drop=True, inplace=True)
df_arsenal.set_index(['performanceId', 'year'], drop=True, inplace=True)

#Option 1 concatenate the two dataframes and group by columns.
#result = pd.concat([df_arsenal, df_pole_sud], 1).dropna().mean(axis=1, level=0)


#Option 2 An alternative would be to find the intersecting indices with index.intersection and index with loc -
i = df_arsenal.index.intersection(df_pole_sud.index)
len(i)

```

```python
df_arsenal = df_arsenal.loc[i]
df_arsenal.sort_values(by=['year'])
```

```python
df_pole_sud = df_pole_sud.loc[i]
df_pole_sud.sort_values(by=['year'])
```
